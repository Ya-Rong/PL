# Step1：匯入資料

[政府公開資料庫](https://data.gov.tw/dataset/7777)

```python
import pandas as pd
data = pd.read_csv("Scenic_Spot.csv") # index_col=0 強制讓第一行變不見
data.head()
```

# Step2：觀察資料

```python
print("資料(資料筆數,欄位數量):",data.shape)
print("資料中所有欄位名稱: ",data.columns)
print("資料中是否有空值: ",data.empty)
```

```python
set(data["Ticketinfo"])
```

```python
set(data["Region"])
```

```python
#篩選資料:單一條件
c1 = (data["Region"] == "臺北市")
d1 = data[c1]

print("共有",len(d1),"筆資料")
d1.head()
```

```python
#篩選資料:多條件
c2 = (data["Region"] == "臺北市") & (data["Town"] == "北投區")
d2 = data[c2]

print("共有",len(d2),"筆資料")
d2.head()
```

```python
#篩選資料中縣市為 臺北市，區域為 北投區 或 大安區
#篩選資料:多條件
c3 = (data["Region"] == "臺北市") & ((data["Town"] == "北投區")|(data["Town"] == "大安區"))
d3 = data[c3]

print("共有",len(d3),"筆資料")
d3.head()
```

```python
#資料名稱.loc[行標籤,列標籤]
loc1 = data.loc[50:55,["Name","Region","Town"]]
loc1
```
