```python
import pandas as pd
from collections import Counter
orgdata = pd.read_csv("bike_road_list_textmining.csv")
orgdata.head()
```

```python
subdata = orgdata[['縣市', '等級', 'seg']]
```

```python
output_series = subdata.groupby(['等級'])['seg'].apply(lambda x: ','.join(x)).reset_index()
```

```python
output_series
```

```python
output_series['seg'] = output_series['seg'].str.replace("', '", " ")
output_series
```

```python
from sklearn.feature_extraction.text import CountVectorizer

docs = output_series['seg'].tolist()
vec = CountVectorizer()
X = vec.fit_transform(docs)
df = pd.DataFrame(X.toarray(), columns=vec.get_feature_names())
df
```

```python
coocc = df.T.dot(df)
coocc
```

```python
import plotly.express as px
fig = px.imshow(coocc)
fig.layout.height = 1000
fig.layout.width = 1000
fig.show()
```

```python
result = coocc.values.flatten()
len(result)
```

```python
df_describe = pd.DataFrame(result)
df_describe.describe()
```

```python
import numpy as np

filter_num_up = 2
filter_num_down = 6
getid = coocc.index
important = pd.DataFrame(columns = getid).T
important['relation'] = ""
dropid = []

for i in range(len(coocc)):
    result = coocc[(coocc.iloc[i] > filter_num_up) & (coocc.iloc[i] < filter_num_down)].index.tolist()
    if(len(result) > 0):
        important['relation'].loc[getid[i]] = coocc[(coocc.iloc[i] > filter_num_up) & (coocc.iloc[i] < filter_num_down)].index.tolist()
        print(important['relation'].loc[getid[i]])
    else:
        dropid.append(i)
        important = important.drop([getid[i]]) 
```
